```python
text="""artificial intelligence is the future artificial intelligence helps humans machine learning is part of artificial intelligence"""
words=text.lower().split()
print(words)
markov_chain={}
for i in range(len(words)-1):
    current_word=words[i]
    next_word=words[i+1]
    if current_word not in markov_chain:
        markov_chain[current_word]=[]
    markov_chain[current_word].append(next_word)
print(markov_chain)
import random
word=random.choice(words)
generated_text=word
for i in range(15):
    if word in markov_chain:
      word=random.choice(markov_chain[word])
      generated_text+=" "+word
    else:
        break
print(generated_text)
```
